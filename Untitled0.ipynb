```python
# covid_data_visualization.py

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
url = 'https://raw.githubusercontent.com/datasets/covid-19/main/data/countries-aggregated.csv'
data = pd.read_csv(url)

# Step 2: Filter data for a few selected countries
countries = ['US', 'India', 'Brazil', 'Russia', 'United Kingdom']
covid_data = data[data['Country'].isin(countries)]

# Step 3: Convert 'Date' column to datetime
covid_data['Date'] = pd.to_datetime(covid_data['Date'])

# Step 4: Group the data by date and country
grouped = covid_data.groupby(['Date', 'Country'])['Confirmed'].sum().unstack()

# Step 5: Plot the confirmed cases
plt.figure(figsize=(10, 6))
grouped.plot(ax=plt.gca())

# Adding labels and title
plt.title('COVID-19 Confirmed Cases Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Confirmed Cases', fontsize=12)
plt.grid(True)
plt.legend(title='Country')
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
